### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

Se podría hacer mediante una función **lambda** o una función normal de python.

Primero lo hacemos con la función **lambda**:

```python
majority = lambda age: age >= 18

df[['age']].apply(majority).head()
```

Ahora lo hacemos mediante una función de python:

```python
def majority(age):
    return age >= 18

df[['age']].apply(majority).head()
```

Finalmente asignamos el resultado a la variable **legal_drinker**:

```python
df['legal_drinker'] = df[['age']].apply(majority)
df.head()
```
